```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda
```

```python
import matplotlib.pyplot as plt
```

#1. Análisis de facturación



```python
ingreso_total_t1 = tienda['Precio'].sum()
print(f'El ingreso total en la tienda 1 es de: {ingreso_total_t1}')
```

```python
ingreso_total_t2 = tienda2['Precio'].sum()
print(f'El ingreso total en la tienda 2 es de: {ingreso_total_t2}')
```

```python
ingreso_total_t3 = tienda3['Precio'].sum()
print(f'El ingreso total en la tienda 3 es de: {ingreso_total_t3}')
```

```python
ingreso_total_t4 = tienda4['Precio'].sum()
print(f'El ingreso total en la tienda 4 es de: {ingreso_total_t4}')
```

```python
ingreso_tiendas = {
    'Tiendas' : ['Tienda1','Tienda2','Tienda3','Tienda4'],
    'Ingreso Tiendas': [ingreso_total_t1,ingreso_total_t2,ingreso_total_t3,ingreso_total_t4]
}

df_ingreso_tiendas = pd.DataFrame(ingreso_tiendas)
df_ingreso_tiendas.set_index('Tiendas',inplace=True)
```

```python
df_ingreso_tiendas
```

```python
#Gráfica de barras del ingreso total por cada tienda

plt.figure(figsize=(10, 5))
df_ingreso_tiendas['Ingreso Tiendas'].plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Ingreso total por cada tienda', fontsize=16)
plt.xlabel('Ingresos Totales')
plt.ylabel('Tienda')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
df_ingreso_tiendas.info()
```

# 3. Calificación promedio de la tienda

```python
calificacion_t2 = tienda2['Calificación'].mean()
print(f'La calificación de los clientes de la tienda 2 es de: {calificacion_t2}')
```

```python
calificacion_t3 = tienda3['Calificación'].mean()
print(f'La calificación de los clientes de la tienda 3 es de: {calificacion_t3}')
```

```python
calificacion_t4 = tienda4['Calificación'].mean()
print(f'La calificación de los clientes de la tienda 4 es de: {calificacion_t4}')
```

```python
calificacion_tiendas = {
    'Tiendas' : ['Tienda1','Tienda2','Tienda3','Tienda4'],
    'Calificación Tiendas': [calificacion_t1,calificacion_t2,calificacion_t3,calificacion_t4]
}

df_calificacion_tiendas = pd.DataFrame(calificacion_tiendas)
df_calificacion_tiendas.set_index('Tiendas',inplace=True)
df_calificacion_tiendas
```

# 5. Envío promedio por tienda

```python
costo_envio_t1 = tienda['Costo de envío'].mean()
print(f'El costo promedio de envío en la tienda 1 es de: {costo_envio_t1}')
```

```python
costo_envio_t2 = tienda2['Costo de envío'].mean()
print(f'El costo promedio de envío en la tienda 2 es de: {costo_envio_t2}')
```

```python
costo_envio_t3 = tienda3['Costo de envío'].mean()
print(f'El costo promedio de envío en la tienda 3 es de: {costo_envio_t3}')
```

```python
costo_envio_t4 = tienda4['Costo de envío'].mean()
print(f'El costo promedio de envío en la tienda 4 es de: {costo_envio_t4}')
```

```python
costo_envio_tiendas = {
    'Tiendas' : ['Tienda1','Tienda2','Tienda3','Tienda4'],
    'Costo Envio Tiendas': [costo_envio_t1,costo_envio_t2,costo_envio_t3,costo_envio_t4]
}

df_costo_envio_tiendas = pd.DataFrame(costo_envio_tiendas)
df_costo_envio_tiendas.set_index('Tiendas',inplace=True)
df_costo_envio_tiendas
```

```python
df_costo_envio_tiendas['Costo Envio Tiendas'] = df_costo_envio_tiendas['Costo Envio Tiendas'].round(2)
df_costo_envio_tiendas
```

```python
#Diagrama de puntos de los costos de envio por cada tienda

plt.figure(figsize=(8, 4))
plt.scatter(df_costo_envio_tiendas.index, df_costo_envio_tiendas['Costo Envio Tiendas'],
            color='#696969', zorder=2)
plt.title('Distribución de Costos de Envío por Tienda')
plt.ylim(0, df_costo_envio_tiendas['Costo Envio Tiendas'].max() * 1.2)
plt.show()
```
